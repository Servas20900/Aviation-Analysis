```python
import pandas as pd
```

```python
# Cargar los dos archivos
df_2025 = pd.read_csv('airport_traffic_2025.csv')
df_2026 = pd.read_csv('airport_traffic_2026.csv')
df_2025.head()
```

### Limpieza y Transformación

```python
# Revisar valores nulos
print('Nulos en 2025:')
print(df_2025.isnull().sum())

print('\nNulos en 2026:')
print(df_2026.isnull().sum())

# Revisar tipos de datos
print('\nTipos de datos 2025:')
print(df_2025.dtypes)

# Estadísticas básicas
df_2025.describe()
```

```python
# Agregar columna de año para identificar cada dataset
df_2025['year'] = 2025
df_2026['year'] = 2026
```

```python
# Unir los dos datasets
df = pd.concat([df_2025, df_2026], ignore_index=True)
```

```python
# Convertir FLT_DATE a formato fecha
df['FLT_DATE'] = pd.to_datetime(df['FLT_DATE'])
```

```python
# Agregar columnas útiles
df['month'] = df['FLT_DATE'].dt.month
df['month_name'] = df['FLT_DATE'].dt.strftime('%B')
df['day_of_week'] = df['FLT_DATE'].dt.day_name()
```

```python
# Eliminar filas con valores nulos en columnas críticas
df = df.dropna(subset=['APT_ICAO', 'FLT_TOT_1'])
```

```python
# Eliminar duplicados
df = df.drop_duplicates()
```

```python
print('Dataset final:', df.shape)
df.head()
```

### Exploración de datos

```python
# Top 20 aeropuertos por tráfico total en 2025
top_airports = df[df['year']==2025].groupby('APT_NAME')['FLT_TOT_1'].sum()
top_airports = top_airports.sort_values(ascending=False).head(20)
print(top_airports)
```

```python
# Tráfico por país
by_country = df[df['year']==2025].groupby('STATE_NAME')['FLT_TOT_1'].sum()
by_country = by_country.sort_values(ascending=False).head(15)
print(by_country)
```

```python
# Tráfico mensual 2025
monthly = df[df['year']==2025].groupby('month')['FLT_TOT_1'].sum()
print(monthly)
```

```python
# Comparación enero 2025 vs enero 2026
jan_2025 = df[(df['year']==2025) & (df['month']==1)]['FLT_TOT_1'].sum()
jan_2026 = df[(df['year']==2026) & (df['month']==1)]['FLT_TOT_1'].sum()
cambio = ((jan_2026 - jan_2025) / jan_2025) * 100
print(f'Enero 2025: {jan_2025:,.0f} vuelos')
print(f'Enero 2026: {jan_2026:,.0f} vuelos')
print(f'Cambio: {cambio:.1f}%')
```

### Exportar el dataset limpio

```python
df.to_csv('aviation_clean.csv', index=False)
print('Archivo guardado: aviation_clean.csv')
print(f'Total de registros: {len(df):,}')
```

## Cargar los datos desde Python a MySQL

```python
# Import
from sqlalchemy import create_engine
```

```python
#Conexión a la db 
engine = create_engine('mysql+mysqlconnector://root:Admin1234!@localhost/aviation_db')
```

```python
# Seleccionar únicamente las columnas necesarias
df_sql = df[[
    'APT_ICAO',
    'APT_NAME',
    'STATE_NAME',
    'FLT_DATE',
    'FLT_DEP_1',
    'FLT_ARR_1',
    'FLT_TOT_1',
    'year',
    'month',
    'month_name',
    'day_of_week'
]].copy()
```

```python

# Renombrar columnas
df_sql.columns = [
    'apt_icao',
    'apt_name',
    'state_name',
    'flt_date',
    'flt_dep',
    'flt_arr',
    'flt_tot',
    'year_num',
    'month_num',
    'month_name',
    'day_of_week'
]
```

```python
# Cargar datos
df_sql.to_sql(
    'flights',
    engine,
    if_exists='replace',
    index=False,
    chunksize=1000
)

print('Datos cargados exitosamente.')
```
